```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import requests
```

# **Getting the data**

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
```

```python
response = requests.get(data_url)
columns = [x.split()[0] for x in response.text.split('\n')[7:21]]
```

```python
data = response.text.split('\n')[22:-1]
```

```python
new_data = []
for row in range(0, len(data), 2):
    row_data = data[row] + data[row+1]
    row_data = list(map(float, row_data.split()))
    new_data.append(row_data)
```

```python
boston_df = pd.DataFrame(new_data, columns=columns)
```

## **EDA**

```python
boston_df = boston_df.rename(columns={'MEDV': 'Price'})
```

```python
boston_df.isnull().sum()
```

```python
import seaborn as sns

correlation_matrix = boston_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Boston Housing Dataset')
plt.show()
```

```python
sns.regplot(x=boston_df['RM'], y=boston_df['Price'])
```

```python
sns.regplot(x=boston_df['LSTAT'], y=boston_df['Price'])
```

```python
sns.regplot(x=boston_df['PTRATIO'], y=boston_df['Price'])
```

## **Independent and dependent features**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = boston_df.iloc[:, :-1]
y = boston_df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## **Model training**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error, mean_absolute_error
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
regression.get_params()
```

```python
reg_pred = regression.predict(X_test)
```

```python
mean_absolute_error(y_test, reg_pred)
```

```python
root_mean_squared_error(y_test, reg_pred)
```

```python
plt.scatter(y_test, reg_pred)
```

```python
r2_score(y_test, reg_pred)
```

```python
residuals = y_test - reg_pred
```

```python
sns.displot(residuals, kind='kde')
```

```python
plt.scatter(reg_pred, residuals)
```

```python
r2_score(reg_pred, y_test)
```
